```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# 1. Data Load & Initial Visualization
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv'
df = pd.read_csv(url)
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

# Plot 1: Original Temperature Trend
plt.figure(figsize=(12, 5))
plt.plot(df['Date'], df['Temp'], color='blue', alpha=0.5)
plt.title('Daily Minimum Temperature Trend')
plt.show()

# 2. Smoothing Analysis
# Plot 2: Moving Average to see the hidden trend
df['Smoothed_Temp'] = df['Temp'].rolling(window=30).mean()
plt.figure(figsize=(12, 5))
plt.plot(df['Date'], df['Temp'], color='blue', alpha=0.3, label='Original')
plt.plot(df['Date'], df['Smoothed_Temp'], color='red', label='30-Day Moving Avg')
plt.legend()
plt.title('Temperature with 30-Day Smoothing')
plt.show()

# 3. Cyclic Feature Engineering
# Plot 3: Unit Circle representation of months
df['Month_sin'] = np.sin(2 * np.pi * df['Month'] / 12)
df['Month_cos'] = np.cos(2 * np.pi * df['Month'] / 12)

plt.figure(figsize=(6,6))
sns.scatterplot(x='Month_sin', y='Month_cos', hue='Month', data=df, palette='viridis')
plt.title('Cyclic Encoding of Months (Unit Circle)')
plt.show()


# 4. Correlation Analysis
# Plot 4: Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df[['Temp', 'Month_sin', 'Month_cos']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation with Cyclic Features')
plt.show()

# 5. Model Evaluation & Residual Plot
train = df[df['Year'] <= 1989]
test = df[df['Year'] >= 1990]
model = LinearRegression().fit(train[['Month_sin', 'Month_cos']], train['Temp'])
y_pred = model.predict(test[['Month_sin', 'Month_cos']])

# Plot 5: Actual vs Predicted
plt.figure(figsize=(12, 5))
plt.plot(test['Date'], test['Temp'], label='Actual', alpha=0.6)
plt.plot(test['Date'], y_pred, label='Predicted', color='red')
plt.legend()
plt.title('Actual vs Predicted Temperature (1990-1991)')
plt.show()

# Plot 6: Residual Distribution
plt.figure(figsize=(8, 4))
sns.histplot(test['Temp'] - y_pred, kde=True, color='purple')
plt.title('Distribution of Prediction Errors (Residuals)')
plt.show()
```
